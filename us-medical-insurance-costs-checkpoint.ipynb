## Analysis of patient data by smoking status

```python
# Analysis of patient data by smoking status. 

import statistics
import matplotlib.pyplot as plt
import csv

patient_data = []
with open('insurance.csv') as insurance_csv:
    insurance_reader = csv.DictReader(insurance_csv, delimiter = ',')
    for row in insurance_reader:
      patient_data.append(row)
      
        
      
      
```

### Total patients in database and total smokers

```python
total_num_of_patients = len(patient_data)
print("There are " + str(total_num_of_patients) + " total patients in the study.")

#Number of smokers compared to total

num_of_smokers = 0
for patient in patient_data:
    if patient['smoker'] == 'yes':
        num_of_smokers += 1
        
print("There are " + str(num_of_smokers) + " total smokers.")

#percentage function

def percentage(whole, amount):
    percentage = amount / whole * 100
    return round(percentage, 2)

    
print("Smokers make up " + str(percentage(1338, 274)) + "% of the total patient population in the study.")
print()

x = ['Total Patients', 'Smokers']
y = [total_num_of_patients, num_of_smokers]
fig = plt.figure(figsize = (10, 6))
plt.title("Fig 1")
plt.xlabel("Total Compared to Smokers")
plt.ylabel("Number of Patients")
plt.bar(x,y)
plt.show()

 #So, out of a patient population of 1,338, there are 274 smokers or 20.48%   
```

### Smoking status by age cohort

```python
#Smokers by age
#First, I want to change values in patient_data that are numbers to integers

for patient in patient_data:
    patient['age'] = int(patient['age'])
    
for patient in patient_data:
    patient['bmi'] = float(patient['bmi'])
    
for patient in patient_data:
    patient['children'] = int(patient['children'])
    
for patient in patient_data:
    patient['charges'] = float(patient['charges'])

    
# Calculate number of patients by age cohort

patients_18_to_29 = 0
patients_30_to_39 = 0
patients_40_to_49 = 0
patients_50_to_59 = 0
patients_60_and_above = 0

for patient in patient_data:
    if patient['age'] >= 18 and patient['age'] <= 29:
        patients_18_to_29 += 1
    elif patient['age'] >= 30 and patient['age'] <= 39:
        patients_30_to_39 += 1
    elif patient['age'] >= 40 and patient['age'] <= 49:
        patients_40_to_49 += 1
    elif patient['age'] >= 50 and patient['age'] <= 59:
        patients_50_to_59 += 1
    else:
        patients_60_and_above += 1
        
# calculate number of smokers in each age cohort

smokers_18_to_29 = 0
smokers_30_to_39 = 0
smokers_40_to_49 = 0
smokers_50_to_59 = 0
smokers_60_and_above = 0

for patient in patient_data:
    if patient['age'] >= 18 and patient['age'] <= 29 and patient['smoker'] == 'yes':
        smokers_18_to_29 += 1
    elif patient['age'] >= 30 and patient['age'] <= 39 and patient['smoker'] == 'yes':
        smokers_30_to_39 += 1
    elif patient['age'] >= 40 and patient['age'] <= 49 and patient['smoker'] == 'yes':
        smokers_40_to_49 += 1
    elif patient['age'] >= 50 and patient['age'] <= 59 and patient['smoker'] == 'yes':
        smokers_50_to_59 += 1
    elif patient['age'] >= 60 and patient['smoker'] == 'yes':
        smokers_60_and_above += 1
    else:
        pass

# Mean, median, mode, min, max for age

patient_ages = []
for patient in patient_data:
    patient_ages.append(patient['age'])
patient_ages.sort()
min_age = min(patient_ages)
max_age = max(patient_ages)
mean_age = statistics.mean(patient_ages)
mode_age = statistics.multimode(patient_ages)
median_age = statistics.median(patient_ages)

rounded_mean_age = round(mean_age, 2)

smoker_ages = []
for patient in patient_data:
    if patient['smoker'] == 'yes':
        smoker_ages.append(patient['age'])
smoker_ages.sort()
min_smoker_age = min(smoker_ages)
max_smoker_age = max(smoker_ages)
mean_smoker_age = statistics.mean(smoker_ages)
mode_smoker_age = statistics.multimode(smoker_ages)
median_smoker_age = statistics.median(smoker_ages)

rounded_mean_smoker_age = round(mean_smoker_age, 2)


print("Patient ages in the study range from " + str(min_age) + " to " + str(max_age) + " years of age.")
print("The average age of patients in the study was " + str(rounded_mean_age) + " years, with a median age of " + 
      str(median_age) + " years. The age occuring most often in the \ndata was " + str(mode_age) + " years.")
print()
print("Smoker ages in the study range from " + str(min_smoker_age) + " to " + str(max_smoker_age) + " years of age.")
print("The average age of smokers in the study was " + str(rounded_mean_smoker_age) + " years, with a median age of " +
     str(median_smoker_age) + " years. The age for smokers occuring most \noften in the data was " + str(mode_smoker_age)
     + " years.")
print()



            
print("There are " + str(patients_18_to_29) + " patients age 18-29, making up " 
      + str(percentage(total_num_of_patients, patients_18_to_29)) + "% of total patient population.")
print("There are " + str(smokers_18_to_29) + " smokers age 18-29, accounting for " + 
      str(percentage(num_of_smokers, smokers_18_to_29)) + "% of the smoking population in this study.")
print(str(percentage(patients_18_to_29, smokers_18_to_29)) + "% of patients age 18-29 are smokers.")
print()
print("There are " + str(patients_30_to_39) + " patients age 30-39, making up " + 
      str(percentage(total_num_of_patients, patients_30_to_39)) + "% of total " + "patient population.")
print("There are " + str(smokers_30_to_39) + " smokers age 30-39, accounting for " + 
      str(percentage(num_of_smokers, smokers_30_to_39)) + "% of the smoking population in this study.")
print(str(percentage(patients_30_to_39, smokers_30_to_39)) + "% of patients age 30-39 are smokers.")
print()
print("There are " + str(patients_40_to_49) + " patients age 40-49, making up " + 
      str(percentage(total_num_of_patients, patients_40_to_49)) + "% of total patient population.")
print("There are " + str(smokers_40_to_49) + " smokers age 40-49, accounting for " + 
      str(percentage(num_of_smokers, smokers_40_to_49)) + "% of the smoking population in this study.")
print(str(percentage(patients_40_to_49, smokers_40_to_49)) + "% of patients age 40-49 are smokers.")
print()
print("There are " + str(patients_50_to_59) + " patients age 50-59 making up " + 
      str(percentage(total_num_of_patients, patients_50_to_59)) + "% of total patient population.")
print("There are " + str(smokers_50_to_59) + " smokers age 50-59, accounting for " + 
      str(percentage(num_of_smokers, smokers_50_to_59)) + "% of the smoking population in this study.")
print(str(percentage(patients_50_to_59, smokers_50_to_59)) + "% of patients age 50-59 are smokers.")
print()
print("There are " + str(patients_60_and_above) + " patients age 60 and above, making up " + 
      str(percentage(total_num_of_patients, patients_60_and_above)) + "% total patient population.")
print("There are " + str(smokers_60_and_above) + " smokers age 60 and above, accounting for " + 
      str(percentage(num_of_smokers, smokers_60_and_above)) + "% of the smoking population in this study.")
print(str(percentage(patients_60_and_above, smokers_60_and_above)) + "% of patients age 60 and above are smokers.")
print()

x = ['Total Patients', '18-29', '30-39', '40-49', '50-59', '60+']
y = [total_num_of_patients, patients_18_to_29, patients_30_to_39, patients_40_to_49, patients_50_to_59, patients_60_and_above]
fig = plt.figure(figsize = (10, 6))
plt.title("Figure 2")
plt.xlabel("Patients by Age")
plt.ylabel("Number of patients")
plt.bar(x,y)
plt.show()
print()

x = ['Total Smokers', '18-29', '30-39', '40-49', '50-59', '60+']
y = [num_of_smokers, smokers_18_to_29, smokers_30_to_39, smokers_40_to_49, smokers_50_to_59, smokers_60_and_above]
fig = plt.figure(figsize = (10, 6))
plt.title("Figure 3")
plt.xlabel("Smoking Cohort by Age")
plt.ylabel("Number of patients")
plt.bar(x,y)
plt.show()
print()

x = ['All Ages', '18-29', '30-39', '40-49', '50-59', '60+']
y = [percentage(total_num_of_patients, num_of_smokers), percentage(patients_18_to_29, smokers_18_to_29),
    percentage(patients_30_to_39, smokers_30_to_39), percentage(patients_40_to_49, smokers_40_to_49),
    percentage(patients_50_to_59, smokers_50_to_59), percentage(patients_60_and_above, smokers_60_and_above)]
fig = plt.figure(figsize = (10, 6))
plt.title("Figure 4")
plt.xlabel("Age category")
plt.ylabel("Percent who smoke")
plt.bar(x,y)
plt.show() 
print()


        
```

### Smoking by BMI

```python
# Breakdown of smoking by BMI, first breakdown of patient bmi by category

underweight = 0
healthy_weight = 0
overweight = 0
obese_1 = 0
obese_2 = 0
obese_3 = 0

for patient in patient_data:
    if patient['bmi'] < 18.5:
        underweight += 1
    elif patient['bmi'] >= 18.5 and patient['bmi'] <= 24.9:
        healthy_weight += 1
    elif patient['bmi'] >= 25 and patient['bmi'] <= 29.9:
        overweight += 1
    elif patient['bmi'] >= 30 and patient['bmi'] <= 34.9:
        obese_1 += 1
    elif patient['bmi'] >= 35 and patient['bmi'] <= 39.9:
        obese_2 += 1
    else:
        obese_3 += 1
        
overweight_and_above = overweight + obese_1 + obese_2 + obese_3 

# Smokers by bmi category

bmi_list = []

for patient in patient_data:
    bmi_list.append(patient['bmi'])

min_bmi = min(bmi_list)
max_bmi = max(bmi_list)
mean_bmi = statistics.mean(bmi_list)
median_bmi = statistics.median(bmi_list)

rounded_mean_bmi = round(mean_bmi, 2)

smoker_bmi_list = []

for patient in patient_data:
    if patient['smoker'] == 'yes':
        smoker_bmi_list.append(patient['bmi'])

smoker_min_bmi = min(smoker_bmi_list)
smoker_max_bmi = max(smoker_bmi_list)
smoker_mean_bmi = statistics.mean(smoker_bmi_list)
rounded_smoker_mean_bmi = round(smoker_mean_bmi, 2)
smoker_median_bmi = statistics.median(smoker_bmi_list)
rounded_smoker_median_bmi = round(smoker_median_bmi, 2)

underweight_smokers = 0
healthy_weight_smokers = 0
overweight_smokers = 0
obese_1_smokers = 0
obese_2_smokers = 0
obese_3_smokers = 0

for patient in patient_data:
    if patient['bmi'] < 18.5 and patient['smoker'] == 'yes':
        underweight_smokers += 1
    elif patient['bmi'] >= 18.5 and patient['bmi'] <= 24.999 and patient['smoker'] == 'yes':
        healthy_weight_smokers += 1
    elif patient['bmi'] >= 25 and patient['bmi'] <= 29.999 and patient['smoker'] == 'yes':
        overweight_smokers += 1
    elif patient['bmi'] >= 30 and patient['bmi'] <= 34.999 and patient['smoker'] == 'yes':
        obese_1_smokers += 1
    elif patient['bmi'] >= 35 and patient['bmi'] <= 39.999 and patient['smoker'] == 'yes':
        obese_2_smokers += 1
    elif patient['bmi'] >= 40 and patient['smoker'] == 'yes':
        obese_3_smokers += 1
    else:
        pass
    
overweight_and_above_smokers = overweight_smokers + obese_1_smokers + obese_2_smokers + obese_3_smokers

print("Patient BMI in the study ranged from " + str(min_bmi) + " to " + str(max_bmi) + ".")
print("The average BMI was " + str(rounded_mean_bmi) + " and the median BMI was " + str(median_bmi) + ".")
print()
print("The BMI for smokers in the study ranged from " + str(smoker_min_bmi) + " to " + str(smoker_max_bmi) + ".")
print("The average BMI for smokers was " + str(rounded_smoker_mean_bmi) + " and the median smoker BMI was " + 
     str(rounded_smoker_median_bmi) + ".")
print()
print(str(underweight) + " patients, making up " + str(percentage(total_num_of_patients, underweight)) + "% of total patients " 
     + "were underweight with a BMI less than 18.5.")
print("Of the underweight cohort, " + str(underweight_smokers) + " were smokers and made up " + 
      str(percentage(num_of_smokers, underweight_smokers)) + "% of the study's smoking population.")
print()
print(str(healthy_weight) + " patients (" + str(percentage(total_num_of_patients, healthy_weight)) + "%)" + 
      " were at a healthy weight with a BMI between 18.5 and 24.99.")
print("Of the healthy weight patients, " + str(healthy_weight_smokers) + " were smokers and made up " +
     str(percentage(num_of_smokers, healthy_weight_smokers)) + "% of the smoking population.")
print()
print(str(overweight_and_above) + " patients were classified as overweight or obese with BMI's of at least 25. This group " +
      " made up " + str(percentage(total_num_of_patients, overweight_and_above)) + "% of the total patient population.")
print("Of the overweight or obese cohort, " + str(overweight_and_above_smokers) + " were smokers, making up " +
     str(percentage(num_of_smokers, overweight_and_above_smokers)) + "% of the smoking population in the study.")
print()
print("The overweight and obese cohort can further be divided as follows:")
print()
print("Patients classified as overweight with BMI's between 25 and 29.999 numbered " + str(overweight) + " and made up " +
     str(percentage(total_num_of_patients, overweight)) + "% of the total patient \npopulation. Of these " + 
      str(overweight_smokers) + " were smokers, making up " + str(percentage(num_of_smokers, overweight_smokers)) +
      "% of the smoking population.")
print()
print("Obese patients can be further divided into three classes. Class 1 obese patients have a BMI between 30 and 34.999. " +
     str(obese_1) + " belong \nto the Class 1 cohort making up " + str(percentage(total_num_of_patients, obese_1)) +
      "% of the total patient population.  Of this group, " + str(obese_1_smokers) + " identified as smokers, making \nup " +
      str(percentage(num_of_smokers, obese_1_smokers)) + "% of the smoking population.")
print()
print("There were " + str(obese_2) + " patients, making up " + str(percentage(total_num_of_patients, obese_2)) + 
      "% of total patients, in the Class 2 obese group with BMI's between 35 and 39.999. " + str(obese_2_smokers) + 
      " \nare smokers and make up " + str(percentage(num_of_smokers, obese_2_smokers)) + 
      "% of the smoking population in the study.")
print()
print("The final category for the obese cohort is Class 3 with BMI's over 40. " + str(obese_3) + 
     " patients are categorized as Class 3 obese making up " + str(percentage(total_num_of_patients, obese_3)) + 
     "% of the patient population. Of these, " + str(obese_3_smokers) + " are smokers and make up " +
      str(percentage(num_of_smokers, obese_3_smokers)) + "% of total smokers.")
print()

#some graphs for visualization

x = ['Total Patients', 'Underweight', 'Healthy Weight', 'Overweight', 'Class 1', 'Class 2', 'Class 3']
y = [total_num_of_patients, underweight, healthy_weight, overweight, obese_1, obese_2, obese_3]
fig = plt.figure(figsize = (10, 6))
plt.title("Figure 8")
plt.xlabel("BMI Category")
plt.ylabel("Number of patients")
plt.bar(x,y)
plt.show()
print()

x = ['Total Patients', 'Underweight', 'Healthy Weight', 'Overweight & Obese']
y = [total_num_of_patients, underweight, healthy_weight, overweight_and_above]
fig = plt.figure(figsize = (10, 6))
plt.title("Figure 9")
plt.xlabel("BMI Category")
plt.ylabel("Number of patients")
plt.bar(x,y)
plt.show()
print()

x = ['Total Smokers', 'Underweight', 'Healthy Weight', 'Overweight', 'Class 1', 'Class 2', 'Class 3']
y = [num_of_smokers, underweight_smokers, healthy_weight_smokers, overweight_smokers, obese_1_smokers, obese_2_smokers,
    obese_3_smokers]
fig = plt.figure(figsize = (10, 6))
plt.title("Figure 10")
plt.xlabel("Smoker BMI Category")
plt.ylabel("Number of patients")
plt.bar(x,y)
plt.show()
print()

x = ['Total Smokers', 'Underweight', 'Healthy Weight', 'Overweight & Obese']
y = [num_of_smokers, underweight_smokers, healthy_weight_smokers, overweight_and_above_smokers]
fig = plt.figure(figsize = (10, 6))
plt.title("Figure 11")
plt.xlabel("Smoker BMI Category")
plt.ylabel("Number of patients")
plt.bar(x,y)
plt.show()
print()

x = ['All BMIs', 'Underweight', 'Healthy Weight', 'Overweight', 'Class 1', 'Class 2', 'Class 3']
y = [percentage(total_num_of_patients, num_of_smokers), percentage(underweight, underweight_smokers),
    percentage(healthy_weight, healthy_weight_smokers), percentage(overweight, overweight_smokers),
    percentage(obese_1, obese_1_smokers), percentage(obese_2, obese_2_smokers), percentage(obese_3, obese_3_smokers)]
fig = plt.figure(figsize = (10, 6))
plt.title("Figure 12")
plt.xlabel("BMI Category")
plt.ylabel("Percent who smoke")
plt.bar(x,y)
plt.show()
print()



```

### Smoking by number of children
